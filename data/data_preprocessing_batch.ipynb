```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('6000_all_categories_questions_with_excerpts.csv')
df
```

```python
from sentence_transformers import SentenceTransformer

# Load the model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Get the excerpts as a list
texts = df['wikipedia_excerpt'].astype(str).tolist()

# Compute embeddings
embeddings = model.encode(texts, show_progress_bar=True)
```

```python
df['embedding'] = list(embeddings)
```

```python
df
```

```python
# Drop columns that are not needed
df = df.drop(columns=['A', 'B', 'C', 'D', 'E', 'answer'])
df
```

```python
df.to_csv('6000_all_categories_questions_with_excerpts_embeddings.csv', index=True)
```
